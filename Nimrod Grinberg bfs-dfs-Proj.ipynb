```python
#Class Node description#

class Node:

    def __init__(self,name):
        self.neighbors = {}
        self.name = name
        self.number_of_neighbors = 0

    def __str__(self):
        print('node name is: {} it has {} neighbors which are:'.format(self.name,self.number_of_neighbors))
        #for neighbor in self.neighbors:
        print(self.neighbors)
        return  'end of node: {}'.format(self.name)

    def __len__(self):  # returns the number of neigbhors of the node
      # num_neihbors = len(self.neighbors.keys())
       return self.number_of_neighbors

    def __contains__(self, item): # returns whether item is a name of a neighbor of self.
        if item in self.neighbors:
            return True
        else:
            return False

    def __getitem__(self, key): # returns the weight of the neighbor named key.
        if key not in self.neighbors:
            return None
        else:
            weight = self.neighbors[key]
            return weight

    def __eq__(self, other): # based on the name attribute
        return self.name == other.name

    def __ne__(self, other):
        return self.name != other.name

    def is_neighbor(self, name): # equivalent to __contains__().
        return name in self.neighbors.keys()

    def update(self, name, weight): # adds name as a neighbor of self. but if weight already exists then it updated it to the bigger one
        if name not in self.neighbors.keys():
            self.neighbors[name] = weight
            self.number_of_neighbors += 1
        else:
            if self.neighbors[name] != weight:
                if self.neighbors[name] < weight:
                    self.neighbors[name] = weight

    def remove_neighbor(self, name): # removes name from being a neighbor of self.
        if name in self.neighbors:
            del self.neighbors[name]
            self.number_of_neighbors -= 1
        else:
            pass

    def is_isolated(self): # returns True if self has no neighbors
        if self.number_of_neighbors == 0:
            return True
        else:
            return False

```

```python
# Class Graph settings #

class Graph():
    def __init__(self, name, nodes=[]):
        self.name = name
        self.number_of_nodes = 0
        self.nodes = {}
        for node_name in nodes:
            self.nodes[node_name.name] = node_name
            self.number_of_nodes += 1

    def __str__(self):
        print('graph {} includes the following nodes:'.format(self.name))
        for node,neigh in self.nodes.items():
            print(neigh)
        return 'end of graph {}'.format(self.name)
       
    def __len__(self):  # returns the number of nodes in the graph
        return self.number_of_nodes

    def __contains__(self,
                     key):  # returns True in two cases: (1) If key is a string, then if a node called key is in self, and (2) If key is a Node, then if a node with the same name is in self.
        if isinstance(key, str) and key in self.nodes:
            return True
        elif isinstance(key, Node) and key in self.nodes:
            return True
        else:
            return False

    def __getitem__(self, name):  # returns the Node object whose name is name
        if name in self:
            return self.nodes[name].name
        else:
            raise KeyError('this name is not in the graph')

    def __add__(self, other):  # returns a new Graph object that includes all the nodes and edges of self and other
        new_list = []
        for value in self.nodes.values():  # here we add the nodes to the list
            new_list.append(value)
        for value in other.nodes.values():  # here we check if a node already been added
            if value not in new_list:
                new_list.append(value)
            else:  # if node of other is already in the new list , this means we need to try and update them
                self.update(value)
        return Graph(self.name+other.name, new_list)

    def update(self, node):  # adds a new node to the graph
        if node.name not in self.nodes.keys():
            self.nodes[node.name] = node
            self.number_of_nodes += 1
        else:
            for key, value in node.neighbors.items():
                self.nodes[node.name].update(key, value)

    def remove_node(self, name):  # removes the node name from self.
        if name in self.nodes.keys():
            del self.nodes[name]
        else:
            pass

    def is_edge(self, frm_name, to_name):  # returns True if to_name is a neighbor of frm_name
        if self.nodes[frm_name].is_neighbor(to_name):
            return True
        else:
            return False


    def add_edge(self, frm_name, to_name, weight):  # adds an edge making to_name a neighbor of frm_name.
        # i need to go to the node name (from name) and add to him a new neighbor (to name) with weight
        if frm_name in self.nodes.keys():
            self.nodes[frm_name].update(to_name, weight)
        else:
            pass

    def remove_edge(self, frm_name, to_name):  # removes to_name from being a neighbor of frm_name.
        if frm_name in self.nodes.keys():
            self.nodes[frm_name].remove_neighbor(to_name)
        else:
            pass

    def get_edge_weight(self, frm_name, to_name):  # returns the weight of the edge between frm_name and to_name.
        if frm_name in self.nodes.keys():
            if self.nodes[frm_name].is_neighbor(to_name):
                return self.nodes[frm_name].neighbors[to_name]
        else:
            pass

       
    def is_reachable(self, frm_name, to_name):
        isolated = []

        if frm_name == to_name:
            return True
        else:
            if frm_name in self.nodes.keys():
                # keep track of all visited nodes
                explored = []
                queue = [frm_name]
                while queue:
                    name = queue.pop()
                    if name not in explored:
                        explored.append((name))
                    else:
                        continue
                    if name == to_name:
                        return True
                    else:
                        if self.nodes[name].is_isolated():
                            isolated.append(name)
                            explored.clear()
                            for i in isolated:
                                explored.append(i)
                            continue
                        for neighbor in self.nodes[name].neighbors.keys():
                            if neighbor == to_name:
                                return True
                            else:
                                if neighbor not in explored:
                                    queue.append(neighbor)


                return False
            else:
                return False

    def get_path_weight(self,path):  # returns the total weight of the given path, where path is an iterable of nodes’ names. # we get a list of nodes names and i check if one node is a nieghbor of the next one and get the weight , that way i calculated the total weight of the path
        if not self.is_reachable(path[0], path[-1]) or len(path) == 0:
            return None
        else:
            total_weight = 0
            for i in range(len(path)-1):
                total_weight += self.get_edge_weight(path[i], path[i + 1])
        return total_weight



    def find_all_paths(self, start, end, path=[]):
        # Find all possible paths to every start and end points
        path = path + [start]
        if start == end:
            return [path]
        if start not in self.nodes.keys():
            return []
        paths = []
        for vertex in self.nodes[start].neighbors.keys():
            if vertex not in path:
                extended_paths = self.find_all_paths(vertex, end, path)
                for p in extended_paths:
                    paths.append(p)
        return paths

    def find_shortest_path(self, frm_name, to_name):
        # use recursion to go over all relevant paths
        paths = self.find_all_paths(frm_name, to_name)
        Path_dict = {}
        for i in paths:
            Path_dict[str(i)] = self.get_path_weight(i)
        if len(Path_dict):
            ans = min(Path_dict.items(), key=lambda x: x[1])
        else:
            ans = None
        return ans

```

```python
# Part III – Non-directional graph #


class NonDirectionalGraph(Graph):
    def __init__(self, name, nodes = []):
        Graph.__init__(self, name,nodes = [])

        self.nodes = {}
        for node in nodes:
            self.nodes[node.name] = node

        self.non_directional = {}
        for node in nodes:
            if node.name not in self.non_directional.keys():
                self.non_directional[node.name] = []
            for key in node.neighbors.keys():
                self.non_directional[node.name].append(key)
                if key not in self.non_directional.keys():
                    self.non_directional[key] = []
                self.non_directional[key].append(node.name)


    def __add__(self, other):
        nondirect_new = NonDirectionalGraph(self.name+other.name, [])
        for key,value in (self.nodes,other.nodes).items():
            nondirect_new.update(value)
        return nondirect_new


    def update(self,node):
        if node.name not in self.non_directional.keys():
            self.non_directional[node.name] = []
        for key in node.neighbors.keys():
            self.non_directional[node.name].append(key)
            if key not in self.non_directional.keys():
                self.non_directional[key] = []
            self.non_directional[key].append(node.name)

        if node.name not in self.nodes.keys():
            self.nodes[node.name] = node

    def remove_edge(self, frm_name, to_name):
        if frm_name in self.non_directional and to_name in self.non_directional:
            self.non_directional[frm_name].remove(to_name)
            self.non_directional[to_name].remove(frm_name)
        else:
            print('the nodes are not in the graph')

    def friend_suggestion(self, friend_name):
        friends_in_common = {}

        for friend in self.non_directional[friend_name]:
            for node in self.non_directional.keys():
                if friend in self.non_directional[node] and node != friend_name and node not in self.non_directional[friend_name]:
                    friends_in_common[node] = friends_in_common.get(node,0) + 1

        return max(friends_in_common,key =lambda i : friends_in_common[i])

```
